## Text Preprocessing

```python
import re

def remove_html(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'', text)

df['review'] = df['review'].apply(remove_html)
```

```python
df
```

```python
# Lower
df['review'] = df['review'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
import nltk

stop_words = stopwords.words('english')
```

```python
df['review'] = df['review'].apply(lambda x: [item for item in x.split() if item not in stopwords.words('english')]).apply(lambda x: " ".join(x))
```

```python
df['review'][0]
```

```python
X = df.iloc[:,:-1]
y = df['sentiment']
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y = encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
# BOW
from sklearn.feature_extraction.text import CountVectorizer  # Import CountVectorizer from the correct submodule

cv = CountVectorizer()
```

```python
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()
```

```python
X_train_bow
```

```python
from sklearn.naive_bayes import GaussianNB

bayes = GaussianNB()
```

```python
bayes.fit(X_train_bow, y_train)
```

```python
y_pred = bayes.predict(X_test_bow)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(X_train_bow,y_train)

y_pred = rf.predict(X_test_bow)

accuracy_score(y_test,y_pred)
```

```python
cv = CountVectorizer(max_features=3000)
```

```python
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()
```

```python
rf.fit(X_train_bow,y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```

## N Grams
